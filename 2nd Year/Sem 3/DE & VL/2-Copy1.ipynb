```python
import pandas as pd

# Load JSON or CSV data into a DataFrame

# Loading JSON file
json_file_path = 'your_file.json'
df_json = pd.read_json(json_file_path)

# Loading CSV file
csv_file_path = 'your_file.csv'
df_csv = pd.read_csv(csv_file_path)

# Display the first few and last few rows of the DataFrame

# Displaying first few rows using head()
print("First few rows of the JSON DataFrame:")
print(df_json.head())

print("\nFirst few rows of the CSV DataFrame:")
print(df_csv.head())

# Displaying last few rows using tail()
print("\nLast few rows of the JSON DataFrame:")
print(df_json.tail())

print("\nLast few rows of the CSV DataFrame:")
print(df_csv.tail())


# Get DataFrame structure and type info
print("\nJSON DataFrame Info:")
print(df_json.info())

print("\nCSV DataFrame Info:")
print(df_csv.info())

# Get descriptive statistics
print("\nDescriptive statistics for JSON DataFrame:")
print(df_json.describe())

print("\nDescriptive statistics for CSV DataFrame:")
print(df_csv.describe())

# Convert DataFrame to string for better readability
print("\nJSON DataFrame as string:")
print(df_json.to_string(max_rows=5))  # Limiting to 5 rows for display

print("\nCSV DataFrame as string:")
print(df_csv.to_string(max_rows=5))  # Limiting to 5 rows for display

# Step 5: Data Cleaning

# Handling missing values
# Replace NaN with a specific value
df_json_cleaned = df_json.fillna(0)  # Replace NaN with 0
df_csv_cleaned = df_csv.fillna('Unknown')  # Replace NaN with 'Unknown'

print("\nCleaned JSON DataFrame (NaN replaced with 0):")
print(df_json_cleaned.head())

print("\nCleaned CSV DataFrame (NaN replaced with 'Unknown'):")
print(df_csv_cleaned.head())

# Dropping rows with missing values
df_json_dropped = df_json.dropna()  # Drop rows with any NaN values
df_csv_dropped = df_csv.dropna()

print("\nJSON DataFrame after dropping rows with missing values:")
print(df_json_dropped.head())

print("\nCSV DataFrame after dropping rows with missing values:")
print(df_csv_dropped.head())

# Dropping duplicate rows
df_json_no_duplicates = df_json.drop_duplicates()
df_csv_no_duplicates = df_csv.drop_duplicates()

print("\nJSON DataFrame after dropping duplicate rows:")
print(df_json_no_duplicates.head())

print("\nCSV DataFrame after dropping duplicate rows:")
print(df_csv_no_duplicates.head())

# Step 6: Save the cleaned DataFrame back to a JSON or CSV file

# Save cleaned JSON DataFrame to file
df_json_cleaned.to_json('cleaned_data.json', orient='records')

# Save cleaned CSV DataFrame to file
df_csv_cleaned.to_csv('cleaned_data.csv', index=False)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"/home/pranali/SYprogramming/DVE/dataset/weight-height.csv")

# Select the first 20 rows
df1 = df.head(20)
print(df1.to_string())

print("------------------------------------")
print("Describe")
print("------------------------------------")

# Calculate the 25th and 50th percentiles for the 'Height' column
Q1 = df['Height'].quantile(0.25)
Q3 = df['Height'].quantile(0.50)

# Calculate IQR
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[(df['Height'] < lower_limit) | (df['Height'] > upper_limit)]
print("Outliers:")
print(outliers)
```
