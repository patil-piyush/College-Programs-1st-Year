```python
import pandas as pd

# Load JSON or CSV data into a DataFrame

# Loading JSON file
json_file_path = 'your_file.json'
df_json = pd.read_json(json_file_path)

# Loading CSV file
csv_file_path = 'your_file.csv'
df_csv = pd.read_csv(csv_file_path)

# Display the first few and last few rows of the DataFrame

# Displaying first few rows using head()
print("First few rows of the JSON DataFrame:")
print(df_json.head())

print("\nFirst few rows of the CSV DataFrame:")
print(df_csv.head())

# Displaying last few rows using tail()
print("\nLast few rows of the JSON DataFrame:")
print(df_json.tail())

print("\nLast few rows of the CSV DataFrame:")
print(df_csv.tail())


# Get DataFrame structure and type info
print("\nJSON DataFrame Info:")
print(df_json.info())

print("\nCSV DataFrame Info:")
print(df_csv.info())

# Get descriptive statistics
print("\nDescriptive statistics for JSON DataFrame:")
print(df_json.describe())

print("\nDescriptive statistics for CSV DataFrame:")
print(df_csv.describe())

# Convert DataFrame to string for better readability
print("\nJSON DataFrame as string:")
print(df_json.to_string(max_rows=5))  # Limiting to 5 rows for display

print("\nCSV DataFrame as string:")
print(df_csv.to_string(max_rows=5))  # Limiting to 5 rows for display


```
