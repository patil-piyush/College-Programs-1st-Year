```python
import pandas as pd
import matplotlib.pyplot as plt

car = ["Jeep","BMW","Audi","Mercedes","Tesla"]
weight = [0.42,1.42,2.0,3.45,5]

data = {'Car':car,'Weight':weight}
df = pd.DataFrame(data)

df.plot(x='Car',y='Weight',kind='line',marker='o')
plt.xlabel('Car')
plt.ylabel('Weight')
plt.title('Car Weights')
plt.show()
```
