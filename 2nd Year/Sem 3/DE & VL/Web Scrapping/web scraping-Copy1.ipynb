```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup as bs

# making the GET request
web = rq.get('https://en.wikipedia.org/wiki/Python_(programming_language)')

# check the status code for response received (200 indicates success)
print(web)

# parse the content using BeautifulSoup
soup = bs(web.content, 'html.parser')

# finding all the tables in the page
table = soup.find_all('table')

# Accessing the first table 
table_all = table[1]

# Extracting table headers
headers = []
for th in table_all.find_all('th'):
    headers.append(th.text.strip())

# Extracting table rows
rows = []
for tr in table_all.find_all('tr'):
    row = [td.text.strip() for td in tr.find_all('td')]
    rows.append(row)

# Creating a DataFrame using pandas
df = pd.DataFrame(rows, columns=headers) 

# Print the DataFrame
print(df)
```
